```python
import requests, bs4
res = requests.get('https://www.premierleague.com/clubs')
```

```python
res.raise_for_status()
```

```python
soup = bs4.BeautifulSoup(res.text, 'html.parser')
```

```python
soup
```

```python
clubs = soup.select('.clubName')
clubs = [club.getText() for club in clubs]
clubs
```

```python
clubLinks = soup.select('.indexItem')
clubLinks = [clubLink.get('href') for clubLink in clubLinks]
clubLinks
```

```python
rootUrl = 'https://www.premierleague.com'
names, clubs, numbers, nationalities, positions, dobs, heights, weights, firstSeasoninPL, yearsinPL = ([] for i in range(10))
for link in clubLinks:
    clubUrl = rootUrl + link.replace('overview','squad')
    res = requests.get(clubUrl)
    soup = bs4.BeautifulSoup(res.text,'html.parser')
    print(f'Accessing {clubUrl}')
    players = soup.select('.playerOverviewCard')
    players = [player.get('href') for player in players]
    for player in players:
        playerUrl = rootUrl + player
        res = requests.get(playerUrl)
        soup = bs4.BeautifulSoup(res.text,'html.parser')
        print(f'----Accessing {playerUrl}')
        clubs.append(soup.select('.info a')[0].getText().strip())
        position = soup.select('.sideWidget .info')[-1].getText()
        positions.append(position)
        try:
            numbers.append(soup.select('.number')[0].getText())
        except:
            numbers.append('NA')
        names.append(soup.select('h1 .name')[0].getText())
        try:
            nationalities.append(soup.select('.playerCountry')[0].getText())
        except:
            nationalities.append('NA')
        infos = soup.select('li .info')[1:]
        if infos:
            dob = infos[0].getText().strip()
        else:
            dob = 'NA'
        if len(infos) > 1:
            height = infos[1].getText()
        else:
            height = 'NA'
        if len(infos) > 2:
            weight = infos[2].getText()
        else:
            weight = 'NA'
        dobs.append(dob.strip())
        heights.append(height)
        weights.append(weight)
        seasons = soup.select('.season')[1:]
        yearsinPL.append(len(seasons))
        firstSeasoninPL.append(seasons[-1].getText())
```

```python
import pandas as pd
df = pd.DataFrame([names, clubs, numbers, nationalities, positions, dobs, heights, weights, firstSeasoninPL, yearsinPL]).T
df.columns = ['Name', 'Club','Jersey No.', 'Nationality', 'Position','Date of Birth', 'Height', 'Weight', 'First Season in PL', 'Years in PL']
```

```python
df.to_csv('pl.csv')
```
